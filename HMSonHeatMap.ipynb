```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
#Read in the data
df = pd.read_csv('Champions-League-Finals-19.csv')

df = df[(df['player_name'] == 'Heung-Min Son')]
df = df.reset_index()

df[['player_name', 'event_type','location_0']]
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#Pitch Template
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

pitch.draw(ax=ax)

#Create the heatmap
kde = sns.kdeplot(
        df['location_0'],
        df['location_1'],
        shade = True,
        shade_lowest=False,
        alpha=.5,
        n_levels=10,
        cmap = 'magma'
)


plt.title('Heung-Min Sons Heat Map 2019 Champions League Final',color='white',size=20)
plt.figtext(.47,.855,'---------->',fontsize=20,ha='left',color='white')


plt.show()
```
